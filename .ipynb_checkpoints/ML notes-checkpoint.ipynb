$RSS(w_0,w_1) = \sum\limits{i}{N} \big(y_i-(w_0+w_1 x_i )\big)^2$

$\hat{w}$ = predicted change in output per unit change in input

## Hill Climbing
Algorithm
$\max_g$ while not converged ($|\frac{dg}{dw}|>\epsilon$), 

$w^(t+1) \leftarrow w^{(t)} - \eta \frac{dg}{dw}\big|_{w^(t)}$

$\rvert_{a=b}$

step size $\eta$ can adust, so can be $\eta_t = \frac{\alpha}{t}$ or $\eta_t = \frac{\alpha}{\sqrt{t}}$

$\eta = .1$ is a good start

```python
x = 0


g = lambda w: 5-(w-10)**2

dg_dw = lambda w: 2 *w - 20
y=-1
eta = 0.05
t=0
w = 0
y = g(w)
epsilon = 0

derivative = dg_dw(w)
while abs(derivative)>epsilon and t<1000:#
    t=t+1
    print('w=',w)
    w = w - eta * dg_dw(w) 
    derivative = dg_dw(w)
    print(t,w,derivative)
```

## Gradient Descent

$\nabla (g)w$ = vector of partial derivatives = $[\frac{dg}{dw_0},\frac{dg}{dw_1},...\frac{dg}{dw_p}]$

a p+1 dimensional vector (assumed $w_0=1$)

$\nabla$ (called 'nabla') is gradient


Algorithm
$\max_g$ while not converged:

    $w^{(t+1)} \leftarrow w^{(t)} - \eta \nabla g(w_t)$

steps opposite of gradient

convergence = $||\nabla g(w) || < \epsilon$ 

    magnitude of gradient less than epsilon

Back to RSS:
    
IF $RSS(w_0,w_1) = \sum\limits_{i}^{N} \big(y_i-(w_0+w_1 x_i )\big)^2$
    
remember $\frac{d}{dw}\sum_{i}^{N} g_i(w) = \sum_{i}^{N}\frac{d}{dw}g_i(w)$

$\nabla RSS(w_0,w_1) = [2(y_i-(w_0+w_1 x_1)\cdot(0 - (1 + 0)=-2(y_i-(w_0+w_1 x_i)),-2(y_i-(w_0+w_1 x_i)x_i)]$ 
$\nabla RSS(w_0,w_1) = -2(y_i-w_0-w_1 x_i),-2(y_i-x_i w_0+w_1 x_i^2)]$

### Closed Form Solution

set gradient to zero and compute gradient=0, plugging solution of $w_0$ from $\frac{\partial g}{\partial w_0}=0$ straight into $\frac{\partial g}{\partial w_1}=0$; will then just need to compute a few terms

$w_1 = \frac{\sum x_iy_i - \frac{1}{N}\sum x_i^2 y_i}{\sum x_i^2 - \frac{1}{N}\sum x^2}$

numerator = (sum of X\*Y) - (1/N)\*((sum of X) \* (sum of Y))

denominator = (sum of X^2) - (1/N)\*((sum of X) \* (sum of X))

or (multiplying by $\frac{N}{N}$)

numerator = (mean of X \* Y) - (mean of X)\*(mean of Y)

denominator = (mean of X^2) - (mean of X)\*(mean of X)

$\frac{\sum x y - \frac{1}{n}\sum x \sum y}{\sum x^2 - \frac{1}{n}\sum x \sum x} =
\frac{\frac{\sum x y}{n} - \frac{\sum x}{n}\frac{\sum y}{n}}{\frac{\sum x^2}{n} - \frac{\sum x}{n}\frac{\sum x}{n}}$

### Gradient Descent Solution

$\nabla RSS(w_0,w_1) = \big[ -2\sum\limits_{i=1}^{N} [y_i - (w_0+w_1x+i), -2\sum\limits_{i=1}^{N} [y_i - (w_0+w_1x_i)]x_i \big]$

while not converged:

   $[w_0^{(t+1)},w_1^{(t+1)}]\leftarrow \big[w_0^{(t)}, w_1^{(t)}\big] - \eta\big[ -2\sum\limits_{i=1}^{N} [y_i - (w_0+w_1x+i), -2\sum\limits_{i=1}^{N} [y_i - (w_0+w_1x_i)]x_i \big]$
